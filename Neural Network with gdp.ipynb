```python
import pandas as pd
```

```python
borderdata = pd.read_csv('data/CleanedBorderdata_2008-2018.csv')
gdpdata = pd.read_csv('data/CleanedGdp2008-2018.csv')
gdpdata
```

```python
borderdata.head()
borderdata = borderdata.rename(columns = {'Latitutde':'Latitude'})
borderdata = borderdata[['Port Name','State','Port Code','Date','Measure','Value','Latitude','Longitude','Border']]
mergeddata = 
```

```python
dataset = borderdata.values
dataset
```

```python
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
from sklearn.model_selection import train_test_split
```

```python
#split dataset into test and train
X_train, X_val_and_test, Y_train, Y_val_and_test = train_test_split(X, Y, test_size=0.3)
```

```python
#split dataset into a validation set
X_val, X_test, Y_val, Y_test = train_test_split(X_val_and_test, Y_val_and_test, test_size=0.5)
```

```python
#    X_train (8 input features, 70% of full dataset)
#    X_val (8 input features, 15% of full dataset)
#    X_test (8 input features, 15% of full dataset)
#    Y_train (1 label, 70% of full dataset)
#    Y_val (1 label, 15% of full dataset)
#    Y_test (1 label, 15% of full dataset)
```

```python
print(X_train.shape, X_val.shape, X_test.shape, Y_train.shape, Y_val.shape, Y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential([    Dense(32, activation='relu', input_shape=(8,)),    Dense(32, activation='relu'),    Dense(1, activation='sigmoid'),])
```

```python
model.compile(optimizer='sgd',              loss='binary_crossentropy',              metrics=['accuracy'])
```

```python
hist = model.fit(X_train, Y_train,          batch_size=32, epochs=100,          validation_data=(X_val, Y_val))
```
